```python
covid_india_data.rename(columns = {'Cured':'Recovered'}, inplace = True)
covid_india_data
```

```python
covid_india_data.drop(['ConfirmedIndianNational', 'ConfirmedForeignNational','Sno'], inplace=True, axis=1)
covid_india_data
```

```python
covid_india_data['Date']=pd.to_datetime(covid_india_data.Date)
covid_india_data
```

```python
covid_india_data['year']=pd.DatetimeIndex(covid_india_data.Date).year
```

```python
covid_india_data['month']=pd.DatetimeIndex(covid_india_data.Date).month
```

```python
covid_india_data['date']=pd.DatetimeIndex(covid_india_data.Date).date
```

```python
covid_india_data['weekday']=pd.DatetimeIndex(covid_india_data.Date).weekday
```

```python
covid_india_data
```

```python
covid_india_data.isnull().sum()
```

```python
covid_india_data.rename(columns = {'State/UnionTerritory':'State'}, inplace = True)
covid_india_data
```

```python
covid_india_data['Prev_Confirmed'] = covid_india_data.groupby('State')['Confirmed'].shift(1)
covid_india_data['New Cases'] = covid_india_data['Confirmed'] - covid_india_data['Prev_Confirmed']
covid_india_data
```

```python
covid_india_data['Prev_Recovered'] = covid_india_data.groupby('State')['Recovered'].shift(1)
covid_india_data['New Recovered'] = covid_india_data['Recovered'] - covid_india_data['Prev_Recovered']
covid_india_data
```

```python
covid_india_data['Prev_Deaths'] = covid_india_data.groupby('State')['Deaths'].shift(1)
covid_india_data['New Deaths'] = covid_india_data['Deaths'] - covid_india_data['Prev_Deaths']
covid_india_data
```

```python
covid_india_data.drop('Prev_Confirmed',inplace = True,axis=1)
covid_india_data.drop('Prev_Recovered',inplace = True,axis=1)
covid_india_data.drop('Prev_Deaths',inplace = True,axis=1)
covid_india_data
```

```python
state_cases = covid_india_data.copy()
state_cases
```

```python
state_cases['Active'] = state_cases['Confirmed'] - (state_cases['Deaths']+state_cases['Recovered'])
state_cases
```

```python
state_cases['Active'] = state_cases['Active'].astype(int)
state_cases["Mortality Rate(per 100)"] = np.round(100*state_cases["Deaths"]/state_cases["Confirmed"],2)
state_cases["Recovered Rate(per 100)"] = np.round(100*state_cases["Recovered"]/state_cases["Confirmed"],2)
state_cases
```

```python
state_cases.reset_index(drop=True,inplace=True)
state_cases.tail(10)
```
